```python
# Create a new column in the dataframe called AgeRange that categorises passengers
# into age ranges: "Child" for ages 0-12, "Teen" for ages 13-19, "Adult" for ages 20-59,
# and "Senior" for ages 60 and above.
# Create bins for age ranges
bins = [0, 13, 20, 60, np.inf]
labels = ['Child', 'Teen', 'Adult', 'Senior']

# Add a new column 'AgeRange' to the dataframe by categorizing 'Age' into bins
df['AgeRange'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)
df
```

```python
# Create a bar chart that shows the total number of passengers in each age range.
bins = [0, 13, 20, 60, np.inf]
labels = ['Child', 'Teen', 'Adult', 'Senior']
# Adding a new column 'Age_range' to the dataframe by categorizing 'Age' into bins
df['Age_range'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)

# Grouping the dataframe by 'Age_range' and getting the count of passengers in each age range
grouped_df = df.groupby('Age_range')['PassengerId'].count()

# Resetting the index to convert the series into a dataframe
grouped_df = grouped_df.reset_index()

# Sorting the dataframe by 'Age_range' to display the bars in order of age ranges
grouped_df = grouped_df.sort_values('Age_range')

# Creating a bar chart based on the 'Age_range' and count of passengers
plt.bar(grouped_df['Age_range'], grouped_df['PassengerId'])

# Adding title and labels
plt.title('Number of Passengers in Each Age Range')
plt.xlabel('Age Range')
plt.ylabel('Number of Passengers')

# Showing the plot
plt.show()
```

```python
# Create a scatter plot that shows the relationship between age and fare. Color the
# points based on whether the passenger survived or not.
plt.scatter(df['Age'], df['Fare'], c=df['Survived'])

# Adding title and labels
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')

# Adding a legend for the color coding
plt.legend({0: 'Did not survive', 1: 'Survived'})

# Showing the plot
plt.show()
```
